```python
# criar novas variáveis
dados["vendas_ativo"] = dados["vendas"]/dados["ativo"] #vendas/ativo
dados["salario_ativo"] = dados["salario_ceo"]/dados["ativo"] #salario_ceo/ativo
dados["log_ativo"] = np.log(dados["ativo"]) #logaritmo do ativo
### apresentar na tabela de variáveis
```

```python
from scipy.stats import ttest_ind
```

```python
grupo0 = dados.loc[dados["educ_business"] == 0, "salario_ceo"]
grupo1 = dados.loc[dados["educ_business"] == 1, "salario_ceo"]

resultado = ttest_ind(grupo0, grupo1)
print(resultado)

###  H0: média do grupo 1 é igual a média do grupo 0
```

```python
grupo0 = dados.loc[dados["educ_business"] == 0, "ativo"]
grupo1 = dados.loc[dados["educ_business"] == 1, "ativo"]

resultado = ttest_ind(grupo0, grupo1, alternative="greater")
print(resultado)
###  H0 com alternative greater: média do grupo 1 é maior que a média do grupo 0
```

```python
import pandas as pd
from scipy.stats import ttest_ind
vars_ = ["idade", "salario_ativo", "vendas_ativo", "ativo"]
resultados = []
for v in vars_:
    g0 = dados[dados["educ_business"] == 0][v].dropna()
    g1 = dados[dados["educ_business"] == 1][v].dropna()
    t, p = ttest_ind(g1, g0, equal_var=False)
    resultados.append({
        "variavel": v,
        "media_grupo0": g0.mean(),
        "media_grupo1": g1.mean(),
        "dif": g1.mean() - g0.mean(),
        "t": t,
        "p": p
    })
tabela = pd.DataFrame(resultados).set_index("variavel").round(3)
tabela

### TABELA DE DIFERENÇAS DE MÉDIAS
```

```python
from scipy.stats import pearsonr

vars_ = [
    "idade", "educ_business", "vendas_ativo",
    "salario_ativo", "log_ativo"
]

df_corr = dados[vars_].dropna()

corr = pd.DataFrame(index=vars_, columns=vars_)
pvals = pd.DataFrame(index=vars_, columns=vars_)

for i in vars_:
    for j in vars_:
        r, p = pearsonr(df_corr[i], df_corr[j])
        corr.loc[i, j] = r
        pvals.loc[i, j] = p

def estrelas(p):
    if p < 0.01:
        return "***"
    elif p < 0.05:
        return "**"
    elif p < 0.10:
        return "*"
    else:
        return ""

corr_sig = corr.astype(float).round(3).astype(str)

for i in vars_:
    for j in vars_:
        corr_sig.loc[i, j] += estrelas(float(pvals.loc[i, j]))

corr_sig
```

```python
import statsmodels.api as sm
```

```python
# definindo as variáveis x e y

y = dados["salario_ativo"]
x = dados[["educ_business", "idade", "vendas_ativo", "log_ativo"]]
x = sm.add_constant(x)
```

```python
# treinar o modelo
modelo = sm.OLS(y, x).fit()
```

```python
# mostrar os resultados

print(modelo.summary())
```
