```python
import pandas as pd # essa biblioteca trazem as funções para operar dados (ex. excel)
import numpy as np # essa biblioteca traz algumas funções matemáticas
```

```python
path = 'winequality-red.csv' # "2026220_20332_DADOS_CEO.xlsx"

print(path)
```

```python
#vamos carregar o excel
dados = pd.read_csv(path)
```

```python
import statsmodels.api as sm
```

```python
y = dados['quality']
x = dados[['alcohol', 'fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
       'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
       'pH', 'sulphates']]
x = sm.add_constant(x)
```

```python
modelo = sm.OLS(y, x).fit()
print(modelo.summary())
```

```python
dados['quality'] = dados['quality'].clip(*dados['quality'].quantile([0.01, 0.99]))
dados['alcohol'] = dados['alcohol'].clip(*dados['alcohol'].quantile([0.01, 0.99]))
dados['fixed acidity'] = dados['fixed acidity'].clip(*dados['fixed acidity'].quantile([0.01, 0.99]))
dados['volatile acidity'] = dados['volatile acidity'].clip(*dados['volatile acidity'].quantile([0.01, 0.99]))
dados['citric acid'] = dados['citric acid'].clip(*dados['citric acid'].quantile([0.01, 0.99]))
dados['residual sugar'] = dados['residual sugar'].clip(*dados['residual sugar'].quantile([0.01, 0.99]))
dados['chlorides'] = dados['chlorides'].clip(*dados['chlorides'].quantile([0.01, 0.99]))
dados['free sulfur dioxide'] = dados['free sulfur dioxide'].clip(*dados['free sulfur dioxide'].quantile([0.01, 0.99]))
dados['total sulfur dioxide'] = dados['total sulfur dioxide'].clip(*dados['total sulfur dioxide'].quantile([0.01, 0.99]))
dados['density'] = dados['density'].clip(*dados['density'].quantile([0.01, 0.99]))
dados['pH'] = dados['pH'].clip(*dados['pH'].quantile([0.01, 0.99]))
dados['sulphates'] = dados['sulphates'].clip(*dados['sulphates'].quantile([0.01, 0.99]))
```

```python
dados.describe().round(2)
```

```python
y = dados['quality']
x = dados[['alcohol', 'fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
       'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
       'pH', 'sulphates']]
x = sm.add_constant(x)
modelo = sm.OLS(y, x).fit()
print(modelo.summary())
```
